 # Исследование оттока клиентов телекоммуникационной компании "Теледом"

# 1 Описание проекта

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Наша задача — обучить на этих данных модель для прогноза оттока клиентов.

Описание услуг
Оператор предоставляет два основных типа услуг: 
Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
Также доступны такие услуги:
Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
Выделенная линия технической поддержки (TechSupport);
Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).
Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

Разобьем данные на обучающую, валидационную и тестовую выборки методом train_test_split. Далее будем использовать pipeline с кодировщиками OneHotencoder и скэйлерами StandardScaler, MinMaScaler и RobustSaler.

```python
features = df_final.drop([
    'customer_id', 'end_contract', 'date',
    'end_date', 'type', 'paper_less_billing',
    'payment_method', 'gender', 'senior_citizen',
    'dependents', 'internet_service','online_security',
    'online_backup', 'device_protection', 'tech_support',
    'streaming_tv', 'streaming_movies'], axis=1)

target = df_final['end_contract']

RANDOM_STATE=28102024

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size = 0.25,
    stratify=target, random_state = RANDOM_STATE 
)

```

Для обучения буем использовать 2 модели. Логистическую регрессию и CatBoostClassifier, так как у нас стоит задача бинарной классификации. Для каждой модели создадим отдельный пайплайн.

```python
num_cols = ['days', 'monthly_charges', 'total_charges']
ohe_cols = ['partner', 'multiple_lines']
```

Создадим пайплайн, используя кодировщики масштабирования и кодирования StandardScaler и OneHotEncoder

```python
ohe_pipe_1 = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(strategy='constant', fill_value='no_service')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first')
        )
    ]
)
```

```python
ohe_pipe_1
```

```python
data_preprocessor_1 = ColumnTransformer(
    [
        ('ohe', ohe_pipe_1, ohe_cols),
        ('num', StandardScaler(), num_cols)
    ],
    remainder='passthrough'
) 
```

```python
data_preprocessor_1
```

```python
pipe_final_1 = Pipeline(
    [
        ('preprocessor', data_preprocessor_1),
        ('models', LogisticRegression(random_state=RANDOM_STATE))
    ]
)
```

```python
param_grid_1 = {
     'preprocessor__num': [
         StandardScaler(),
         MinMaxScaler(),
         RobustScaler(),
         'passthrough'   
    ]
}
```

Построим кривую roc_curve для наглядности

```python
fpr, tpr, _ = roc_curve(target_test, (predictions))
plt.plot(fpr, tpr)
plt.title("График зависимости FPR от TPR")
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.show()
```

Значение метрии roc_auc - это доля правильно упорядоченных пар. Значение метрики на графике кривой roc_curve - это площадь под ривой. Можно сказать чем больше эта площадь, тем больше TPR(полнота - доля положительных обьетов, правильно предсказанных положительными и тем меньше (FPR) - доля отрицательных обьетов, неверно предсказанных положительными. Тоесть корректнее гворить о доле верно упорядоченных пар обьектов. Чем больше площадь под кривой, тем больше значение roc_auc и тем меньше ошибается модель. Но так как у нас нет порога классификации(тоесть цены ошибки), мы не можем точно сказать на сколько ошибается модель и не можем посчитать спрогнозировать будущую прибыль или убыток. Но можно посчитать метрику accuracy - долю верно упорядоченных пар или долю правильно классифициремых обьектов.

Далее построим матрицу ошибок confusion_matrix для анализа ошибок первого и второго рода.

```python
confusion = np.zeros((2, 2)).astype(int)
for i in range(2):
    for j in range(2):
        confusion[i][j] = np.sum(np.logical_and(target_test == i, np.round(predictions) == j))
print(confusion)                                               
```

```python
sns.heatmap(confusion, annot=True, fmt='d', cmap='inferno')
```

По условию задачи нам нужно предсказать действие абонента - "разорвет ли он договор или нет". Тоесть по другому нам нужно не допустить, чтобы абонент ушел из компании. Тоесть нам нужно избегать ошибки первого рода - FP(False Positive).
Тут ошибка первого рода или FP = 107. Тоесть в 113 случаях модель не верно предсказывает класс 1, тоесть то, что клиент останется.
